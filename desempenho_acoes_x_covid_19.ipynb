## Análise do efeito da pandemia nas ações 

Análise do desempenhdo de ações e as consequências do fechamento do comercio, baseadas em inferência lógica, lockdown.

Tipo de ações: Empresas de shopping centers do Brasil.

## Bibliotecas a serem importadas

```python
import pandas as pd
from pandas_datareader import data as web
import plotly.graph_objects as go
```

## Definição Stocks

Quatro das maiores empresas de shoppings centers do país — Aliansce Sonae, BR Malls, Multiplan e Iguatemi.

```python
# escolher a ação desejada, acompanhada da terminação .SA
acao1 = 'MULT3.SA'
acao2 = 'BRML3.SA'
acao3 = 'ALSO3.SA'
acao4 = 'IGTA3.SA'
```

## BVMF: MULT3

```python
# criar um DataFrame vazio para receber os valores posteriormente
df1 = pd.DataFrame()

# importar dados para o DataFrame selecionando o período de início da exibição - no caso, 01-01-2019
df1 = web.DataReader(acao1, data_source='yahoo', start='01-01-2019')

# analisando as primeiras entradas do arquivo
df1.head()
```

```python
# definindo as configs
trace1 = {
    'x': df1.index,
    'open': df1.Open,
    'close': df1.Close,
    'high': df1.High,
    'low': df1.Low,
    'type': 'candlestick',
    'name': acao1,
    'showlegend': True
}

data1 = trace1
layout1 = go.Layout()
 
acao1 = go.Figure(data=data1, layout=layout1)
```

```python
# aqui vão algumas atualizações para adicionar a linha de destaque
acao1.update_layout(
    title='Reflexo da pandemia COVID-19 - Multiplan (MULT3)',
    shapes = [dict(
        x0='2020-02-26', x1='2020-02-26', y0=0, y1=1, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2020-02-26', y=0.05, xref='x', yref='paper',
        showarrow=False, xanchor='left', text='Início da Covid')]
)

# exibindo
acao1.show()
```
